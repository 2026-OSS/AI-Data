```python
import glob
import os
from IPython.display import Image, display

DATASET_DIR = "./AI-Picture-Book-Object-Detection-6"
YAML_PATH = f"{DATASET_DIR}/data.yaml"

# 가장 최근 학습된 best.pt 찾기
best_list = sorted(glob.glob("runs/train/*/weights/best.pt"))

if not best_list:
    raise FileNotFoundError("best.pt 파일을 찾지 못했습니다. 먼저 학습을 실행해주세요.")

best_weights = best_list[-1]
print("Best weights:", best_weights)

# 테스트 이미지 폴더 확인
test_source = f"{DATASET_DIR}/test/images"
valid_source = f"{DATASET_DIR}/valid/images"

if os.path.exists(test_source):
    source_path = test_source
    val_task = "test"
else:
    source_path = valid_source
    val_task = "val"

print("Inference source:", source_path)
print("Validation task:", val_task)

# 이미지 추론
!python detect.py \
--weights {best_weights} \
--conf 0.25 \
--source {source_path}

# 추론 결과 확인
results = sorted(glob.glob("runs/detect/exp*/*.jpg"))

print(f"Detected Images: {len(results)}")

for img in results[-10:]:
    display(Image(filename=img))

# Dataset 평가
!python val.py \
--weights {best_weights} \
--data {YAML_PATH} \
--img 640 \
--task {val_task}

# 학습 결과 그래프 표시
result_imgs = sorted(glob.glob("runs/train/*/results.png"))

if result_imgs:
    print("Training Result:", result_imgs[-1])
    display(Image(filename=result_imgs[-1]))
else:
    print("results.png 파일을 찾지 못했습니다.")
```
